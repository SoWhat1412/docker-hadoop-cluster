```python
import findspark
findspark.init()
import pyspark
import random
```

```python
conftfos = pyspark.SparkConf().setAll([('spark.executor.cores','1')])
```

```python
sc = pyspark.SparkContext(appName="Pi", conf=conftfos)
```

```python
RDDread = sc.textFile("file:///usr/local/spark/NOTICE")
```

```python
RDDread.collect()
```

```python
RDDread.count()
```

```python
mappedRDD = RDDread.map(lambda line : line.split(" "))
mappedRDD.take(5)
```

```python
flatmappedRDD = RDDread.flatMap(lambda line : line.split(" "))
flatmappedRDD.take(5)
```

```python
filteredRDD = RDDread.filter(lambda line : "apache" in line.lower())
print(filteredRDD.take(5))
filteredRDD.count()
```

```python
filteredRDD.getNumPartitions()
```

```python
filteredRDD.collect().foreach(print)
```

```python
sc.stop()
```
