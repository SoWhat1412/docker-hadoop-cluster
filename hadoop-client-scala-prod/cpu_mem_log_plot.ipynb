```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('cpu_mem_log.log')
df['date_time_index'] = pd.to_datetime(df['DATETIME'], format='%Y%m%d %H:%M:%S')
df.set_index(['date_time_index'],inplace=True)
df.drop('DATETIME', axis=1, inplace=True)

%matplotlib inline
%matplotlib inline
```

```python
fig = plt.figure(figsize=(12,7))
ax1, ax2, ax3, ax4 = fig.subplots(4,1,sharex=True)


ax1.plot(df.index, df['%CPU'], 'r-')
ax2.plot(df.index, df['%MEM'], 'b-')
ax3.plot(df.index, df['DISK(GB) for /data/tmp/tmp-docker'], 'm-')
ax4.plot(df.index, df['netstat -an | grep :9989 | wc -l'], 'g-')

ax1.set_ylabel('%CPU', color='r')
ax2.set_ylabel('%MEM', color='b')
ax3.set_ylabel('DISK(GB) for /data/tmp/tmp-docker', color='m')
ax4.set_ylabel('netstat -an | grep :9989 | wc -l', color='g')

ax4.set_xlabel('datetime')

plt.show()
```
